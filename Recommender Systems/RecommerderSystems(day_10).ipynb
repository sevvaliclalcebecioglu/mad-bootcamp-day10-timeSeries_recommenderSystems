```python
rating = pd.merge(movies, ratings)

# Film isimleri ve kullanıcı puanlarını tek bir DataFrame'de birleştiriyoruz
```

```python
rating.title.value_counts()

# Her filmin kaç kez oy aldığını sayarak popülerliklerini kontrol ediyoruz
```

```python
import numpy as np

# Sayısal işlemler ve diziler için NumPy kütüphanesini yüklüyoruz
```

```python
mg = rating.groupby('title').agg({'rating': [np.size, np.sum, np.mean]})

# Her film için oy sayısı, toplam puan ve ortalama puanı hesaplayarak özet istatistikler oluşturuyoruz
```

```python
mg

# Film bazlı özet istatistikleri görüntüleyerek oy sayısı, toplam puan ve ortalama puan bilgilerini inceliyoruz
```

```python
pm = mg.sort_values(('rating', 'mean'), ascending=False)

# Filmleri ortalama puana göre azalan sırada sıralayarak en yüksek puanlı filmleri belirliyoruz
```

```python
pm

# Ortalama puana göre sıralanmış film istatistiklerini görüntüleyerek en beğenilen filmleri inceliyoruz
```

```python
gs = pm['rating']['sum'].sum()

# Tüm filmlerin aldığı toplam puan sayısını hesaplıyoruz
```

```python
pm['percentage']=pm['rating']['sum'].div(gs)*100
```

```python
gs
```

```python
pm.head()
```

```python
pm['rank']=pm['percentage'].rank(ascending=False)
```

```python
pm.sort_values(by='rank',ascending=True)
```
