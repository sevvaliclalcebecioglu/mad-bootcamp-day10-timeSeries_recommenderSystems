```python
rating = pd.merge(movies, ratings)

# Film isimleri ve kullanıcı puanlarını tek bir DataFrame'de birleştiriyoruz
```

```python
rating.head()

# Birleştirilmiş DataFrame'in ilk 5 satırını görüntüleyerek kullanıcı-film-puan ilişkisini inceliyoruz
```

```python
# rating=pd.merge(rating,user)
```

```python
# rating.head()
```

```python
rating.title.value_counts()

# Her filmin kaç kez oy aldığını sayarak popülerliklerini kontrol ediyoruz
```

```python
mf = rating.pivot_table(index=['user_id'], columns=['title'], values='rating')

# Kullanıcıları satır, film isimlerini sütun olarak alıp puanları tabloya yerleştiriyoruz (pivot table oluşturuyoruz)
```

```python
mf.head()

# Pivot tablonun ilk 5 satırını görüntüleyerek kullanıcı-film puan yapısını inceliyoruz
```

```python
swr = mf['Star Wars (1977)']

# Pivot tablodan 'Star Wars (1977)' filminin tüm kullanıcılar için puanlarını seçiyoruz
```

```python
swr.value_counts()

# 'Star Wars (1977)' filminin aldığı oyların dağılımını ve her puanın kaç kez verildiğini sayıyoruz
```

```python
mf[['Star Wars (1977)', '101 Dalmatians (1996)']].corr()

# İki filmin kullanıcılar tarafından verilen puanları arasındaki Pearson korelasyonunu hesaplayarak benzerliklerini inceliyoruz
```

```python
sm = mf.corrwith(swr)

# Tüm filmlerin 'Star Wars (1977)' ile kullanıcı puanları bazında korelasyonunu hesaplıyoruz
```

```python
sm.sort_values(ascending=False)

# Bu satır, sm değişkenindeki değerleri büyükten küçüğe doğru sıralar.
```

```python
import numpy as np

# NumPy kütüphanesini sayısal hesaplamalar ve dizi işlemleri için programa dahil eder.
```

```python
mstats = rating.groupby('title').agg({'rating': [np.size, np.mean]})

# Her film başlığı için kaç tane puan verildiğini ve ortalama puanı hesaplar.
```

```python
mstats.head()

# Oluşturulan istatistik tablosunun ilk 5 satırını görüntüler.
```

```python
pm = mstats[mstats['rating']['size'] > 100]

# 100’den fazla oy alan filmleri seçerek daha güvenilir istatistikler elde eder.
```

```python
smdata = pd.DataFrame(sm, columns=['similarity'])

# Benzerlik değerlerini 'similarity' adlı sütunla bir DataFrame yapısına dönüştürür.
```

```python
smdata

# Oluşturulan DataFrame’in içeriğini ekrana yazdırır.
```

```python
pm.columns = pm.columns.get_level_values(0)
df = pm.join(smdata)

# Çok seviyeli sütun yapısını sadeleştirir ve film istatistiklerini benzerlik verileriyle birleştirir.
```
